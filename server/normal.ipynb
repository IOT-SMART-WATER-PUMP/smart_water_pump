```python
import numpy as np
import matplotlib.pyplot as plt

# Normal distribution function with adjusted scaling to maintain same height
def normal_dist_same_height(x, mean, std, ref_std):
    scale_factor = 1 / (ref_std * np.sqrt(2 * np.pi))
    return scale_factor * (1 / (std * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean) / std) ** 2)

if __name__ == '__main__':
    mean = 30
    ref_std = 8  # Reference standard deviation for same height

    # Generate x values for plotting (use linspace for smooth plotting)
    x = np.linspace(0, mean, 1000)

    # First normal distribution with std = 5
    std_1 = 5
    y_1 = [normal_dist_same_height(xi, mean, std_1, ref_std) for xi in x]
    plt.plot(x, y_1, label=f"Std = {std_1}")

    # Second normal distribution with std = 20
    x1 = np.linspace(mean, 4*mean, 1000)
    std_2 = 30
    y_2 = [normal_dist_same_height(xi, mean, std_2, ref_std) for xi in x1]
    y_2 = [y / max(y_2) * max(y_1) for y in y_2]
    plt.plot(x1, y_2, label=f"Std = {std_2}")

    # Adding labels and title
    plt.title(f"Normal Distributions with Same Peak Height (mean={mean})")
    plt.xlabel("x")
    plt.ylabel("Probability Density")
    plt.legend()  # Show legend
    plt.grid(True)  # Show grid
    plt.show()
```
